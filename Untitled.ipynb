# Linear Regression Modeling to Predict House Prices

The purpose of this notebook is to apply linear regression to attempt to predict median house values(MEDV) for educational purposes. A linear model can be defined as y = X * beta + epsilon, where epsilon denotes the residuals of the model, ie. random noise. Linear regression is based on a set of statistical assumptions, which need to be satisfied for it to work properly. The exact set of assumptions tends to vary depending on the source, but the most important ones are as follows: 

## 1. Linearity

The relationship between the explanatory variables X and the target variable y is linear. 

## 2. Linear independence between explanatory variables. 

The explanatory variables X are linearly independent, which means that they are uncorrelated. If this assumption is violated, we say there is multicollinearity in X. This can make it difficult to analyze the effect of individual variables on the model, and it can make X singular and thus prevent us from using the closed form solution. 

## 3. The residuals have zero mean. 

The residuals of the model are cetered at zero. This ensures that the expected value of y is equal to X * beta, ie. there is no systematic bias in the model. 

## 4. The residuals have constant variance. 

If this assumption holds, the residuals are said to be homoschedastic. If it does not hold, the residuals are said to be heteroschedastic. Heteroschedasticity can make the model unstable, as the error rate of the model varies in the input space. 

## 5. The residuals are uncorrelated. 

Violation of this assumption can lead to the least square estimators ineffective.

## (6.) The residuals are normally distributed.

This assumption is not really necessary, but it makes statistical testing less computationally demanding. 



This notebook attempts to effectively test all of these assumptions, address all possible violations of these assumptions and find the best possible linear model for this dataset. 

```python
#Import libraries and define constants

import numpy as np
import pandas as pd
from clean import clean_data
import seaborn as sn
from matplotlib import pyplot as plt

PATH = "data/data.txt"
OUT_PATH = "data/data_clean.csv"
```

```python
#Clean up downloaded dataset

clean_data(PATH, OUT_PATH)
df = pd.read_csv("data/data_clean.csv", names=["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
                                              "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"])
print(df.head)
```

```python
#Examine key univariate statistics of the dataset. 
df.describe()
```

```python
'''Plot correlation of dataset. We can see from the last column that 
LSTAT has a strong negative correlation with MEDV, and RM has a strong 
positive correlation with MEDV. LSTAT and RM are also strongly (negatively) correlated. 
Based on these observations we could expect there to be multicollinearity in X.'''
cor = df.corr()
sn.heatmap(cor)
plt.show()
```

```python
'''
Add extra column of ones to X for the purpose of including the intercept term in the parameter estimates. 
Attempt to produce least square estimates with the closed form solution.
'''
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values.reshape(-1, 1)
X = np.column_stack((np.ones(y.shape), X))

try:
    beta_hat = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    print(beta_hat)
except np.linalg.LinAlgError:
    print("Dependent columns in X.")
```

```python
'''
We were able to obtain the parameter estimates. Let's proceed to obtain target variable estimates
and residual estimates.
'''
y_hat = np.dot(X, beta_hat)
res = y_hat - y
```

```python
'''
Let's start by plotting y_hat against y to have some visual sense of how well the model estimates y. 
The closer this plot follows a line with slope 1, the better the model estimates y. We can see that the 
estimates start to deviate from the ground truth values more significantly at the high end. 
'''
plt.scatter(y_hat, y)
plt.plot([y.min(), y.max()], [y.min(), y.max()], linewidth = 5, c = "red")
plt.xlabel("prediction")
plt.ylabel("ground truth")
plt.show()
```
